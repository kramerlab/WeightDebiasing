```python
import numpy as np
from scipy.special import xlogy
import seaborn as sns
from scipy.stats import entropy
sns.set_style("darkgrid")
import matplotlib.pyplot as plt
```

```python
y_codes = np.array([-1.0, 1.0])
```

```python
predictions = np.linspace(0, 1, 10000)
```

```python
epsilon = np.finfo(float).eps
predictions = np.clip(predictions, a_min=epsilon, a_max=None)
y_coding = y_codes.take(predictions < 0.5)
p_difference =  np.abs(0.5 - predictions)
```

```python
ax = sns.lineplot(x=predictions, y=p_difference)
ax.set_xticks(np.arange(0, 1.1, 0.25))
ax.set_xlabel("Predicted Probability")
plt.savefig("absolute_difference_factored_function.pdf")
```

```python
ax = sns.lineplot(x=predictions, y=np.power(p_difference, 0.75)) 
ax.set_xticks(np.arange(0, 1.1, 0.25))
ax.set_xlabel("Predicted Probability")
plt.savefig("absolute_difference_powered_function.pdf")
```

```python
tmp = y_coding * np.power(p_difference, 0.5)
ax = sns.lineplot(x=predictions, y=tmp) 
ax.set_xticks(np.arange(0, 1.1, 0.25))
ax.set_xlabel("Predicted Probability")
ax.set_ylabel("Weight Update")
plt.savefig("inversed.pdf")
```

```python
tmp = -predictions + 1.5
ax = sns.lineplot(x=predictions, y=tmp) 
ax.set_xticks(np.arange(0, 1.1, 0.25))
ax.set_xlabel("Predicted Probability")
ax.set_ylabel("Weight Update")
plt.savefig("weight_update_function.pdf")
```
